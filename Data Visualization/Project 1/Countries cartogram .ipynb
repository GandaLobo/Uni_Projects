```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import geoplot
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world.plot()
```

```python
world
```

```python
world.set_index('name', inplace = True)
```

```python
world
```

```python
dfc= pd.read_excel('Cases_million.xlsx')
```

```python
dfc
```

```python
dfc.drop(['Unnamed: 2', 'Unnamed: 3','Unnamed: 4'], axis=1, inplace = True)
```

```python
dfc.set_index('Country',inplace=True)
```

```python
dfc
```

```python
w = world.join(dfc)
```

```python
w
```

```python
w.drop(['pop_est','continent','iso_a3','gdp_md_est'], axis = 1,inplace = True)
```

```python
dfworld = gpd.geodataframe.GeoDataFrame(data=w)
```

```python
dfworld
```

```python
dfworld.plot()
```

```python
base = dfworld.plot(column='Cases',missing_kwds={'color': 'lightgrey'},cmap='Reds',figsize=(33.1,46.8)
),legend=True,legend_kwds={'label': "Cases per million",'orientation': "horizontal"});
plt.title("Cases per Million Habitants")
dfworld.boundary.plot(ax=base, color='black')
plt.savefig('map1.pdf')


```

```python
base = dfworld.plot(column='Cases',missing_kwds={'color': 'lightgrey'},cmap='Reds',figsize=(46,33),legend=True,legend_kwds={'label': "Cases per million",'orientation': "horizontal"});
plt.title("Cases per Million Habitants")
dfworld.boundary.plot(ax=base, color='black')
plt.savefig('map4.jpeg')
```
